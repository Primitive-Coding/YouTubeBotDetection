```python
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
import warnings

# Suppress FutureWarnings
warnings.filterwarnings("ignore", category=FutureWarning)
```

```python
# Load pre-trained model
model = SentenceTransformer('paraphrase-MiniLM-L6-v2')
```

# Embeddings Comments


```python
comment1 = "This is a test" 
comment2 = "This is a simple test"

emb1 = model.encode(comment1)
emb2 = model.encode(comment2)

cosine_sim = cosine_similarity([emb1], [emb2])


sim = cosine_sim[0][0] * 100
sim = "{:,.2f}%".format(sim)

print(emb1)
print(emb2)
print(sim) 
```

# Embedding Multiple Comments

```python
comments = [
    "This is the first comment.",
    "Another comment is here.",
    "Yet another comment for comparison.",
    "This is similar to the first comment."
]

embeddings = model.encode(comments)
# Compute the cosine similarity matrix
similarity_matrix = cosine_similarity(embeddings)

n_comments = len(comments)

data = {
    "source": [],
    "target": [],
    "similarity": []
}
for i in range(n_comments):
    for j in range(i + 1, n_comments):  # Ensure we don't compare the same pair twice
        data['source'].append(comments[i])
        data['target'].append(comments[j])
        data['similarity'].append(similarity_matrix[i][j])
        
    


data = pd.DataFrame(data)
# data.sort_values(by="similarity", ascending=False, inplace=True)
# data.reset_index(drop=True, inplace=True)
data
```
